# $\LaTeX$ MATH COMMANDS

# Commands
---

In Jupyter notebooks, the math symbols can be rendered using LaTeX commands, e.g.:

![The LaTeX commands for the Greek Alphabet](images/LaTeX_commands_for_Greek_alphabet.png "The LaTeX commands for the Greek Alphabet")

These are commands that modify the font size:

* \Huge
* \huge
* \LARGE
* \Large
* \large
* \normalsize (default)
* \small
* \footnotesize
* \scriptsize
* \tiny
