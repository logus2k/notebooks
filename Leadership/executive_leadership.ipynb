# EXECUTIVE LEADERSHIP
---

# The Two Principles
---

## Principle 1: Earn Trust

Trust is believing that a person will do what you expect them to do. However, it also tends to mean that the person we trust will act with our best interests in mind. As an executive, it is important that you remember that trust is a key to success. People will follow you during times of uncertainty and will be more willing to accept change when they trust you. So, work to build trust every day.

Earning trust every day is the foundation of successful executive leadership. Whether they say it or not, a question on everyone’s mind is “Why should we follow you?” The best answer is “Because we trust you.”

We trust you to look out for our best interest, to help us succeed, empower us to reach our goals, and do more than we thought we could. To help us be proud of the meaningful work we’ve done. But there’s only one way to gain it and keep it, and that’s to earn it.

Do this everyday and you will build strong trust:

 * __Reliability__: be someone who can be counted on. Show up on time, be prepared, and over-deliver. Make commitments and follow through. Close the loop on all expectations. If sometimes you fall short, do whatever it takes to make it right.

 * __Credibility__: Bring value to the table. Know what you are supposed to know and do what you are supposed to do. Learn all about your business and industry. Step up when you are needed. Make your presence felt and add value.

 * __Connection__: build rapport and relationship with others. Learn something from them or about them every time you interact. Find out what matters to them and help them get it. Listen, understand, and give. 

## Principle 2: Serve Your People

The heart of executive leadership is in two words, “serving others.” Orient your personal compass to serving
others and your impact will be far-reaching.

One of my favorite role models I spoke to in my research is Mike Critelli, the former CEO of postal service giant
Pitney Bowes who led them through an era of extraordinary success. Critelli’s compass was always pointing
toward “serve others,” even in passing conversations. At a sales conference he chatted with an employee and
learned he and his wife were adopting a child. A few weeks later they received a personal letter from Critelli
congratulating them on their new child along with a check for the amount of the new adoption benefit the
company just started offering, thanks to that brief conversation.

The little things matter big; they add up. When your leadership compass is pointed toward “serve others,”
what’s the very best way you can serve your people today, this week, this month, this year?

# The Four Disciplines
---

Like an orchestra conductor, an executive leader does not make the music. They serve the people who do. In two words, the heart of executive leadership is "serving others".

The four disciplines of executive leadership are top priority, integrated sets of practices and actions for achieving success:

## 1. Set Direction

_"Set your course by the stars, not by the lights of every passing ship"_ - Omar Bradley

 * Think strategically _\[Strategic Thinking\]_
 * Take responsibility and be decisive _\[Decisiveness And Courage\]_
 * Lead from “me” to “we”: define the past, present, and future _\[Past, Present And Future\]_
 * Create a shared purpose and compelling vision _\[Purpose And Vision\]_

## 2. Motivate Commitment

_"If you want to build a ship, don't drum up the people to gather wood, divide the work, and give orders. Instead, teach them to yearn for the vast and endless sea"_ - The Little Prince

 * Inspire confidence, even under pressure _\[Manage Your Image\]_
 * Energize and empower people _\[Energize And Empower\]_
 * Encourage personal excellence _\[Encourage Personal Excellence\]_
 * Create collaboration opportunities _\[Create Collaboration Opportunities\]_
 * Develop others and build a talent pipeline _\[Develop Others\]_
 * Communicate to motivate _\[Communicate To Motivate And Inspire\]_
 * Communicate to replicate _\[Communicate To Replicate\]_

## 3. Drive For Results

_"The pessimist complains about the wind; the optimist expects it to change; the realist adjusts the sails"_ - William Arthur Ward

 * Establish priorities and focus _\[Prioritization\]_
 * Ensure invigorating accountability _\[Accountability\]_
 * Influence without authority _\[Influence Without Authority\]_
 * Cultivate creative thinking and innovation _\[Cultivating Creativity\]_
 * Lead large-scale change and shape culture _\[Leading Change\]_

 * Build resilience and resourcefulness _\[Resilience\]_
 * Keep learning _\[Learning\]_
 * Keep connecting _\[Connections\]_

# Strategic Thinking
---

Most people's thinking is dominated by the here and now, but effective executives make a habit of thinking ahead in time and space. They think in a disciplined way about the future, and they think beyond their tasks and team. They consider implications for their organization as a whole, as competitors, the economy, social implications, even global developments. That's strategic thinking, and it reveals problems and possibilities most others miss.

# Decisiveness And Courage
---

"_Leadership  is that splendid misery that involves the daily loss of friends"_ - Thomas Jefferson

After assessing thousands of CEO candidates over the years and examining in detail who succeeded and failed, research revealed a key success factor is acting decisively and courageously.

Here is how to do it:

1. Make clear, justified choices
2. Collect diverse inputs
3. Challenge your assumptions
4. Consult trusted advisors regularly

# Past, Present And Future
---

"_A good leader inspires others with confidence in him; a great leader inspires them with confidence in themselves_" - Franklin Roosevelt

Especially during times of stress or change, it's natural for people's focus on "me" to overcome their sense of "we". So during good times and hard times alike, executives need to draw people together and point them forward with clarity and energy. They need to lead from "me" to "we".

This is the right way to frame the past, present and future:

 * __Past__: What did we accomplished that make us proud?

 "_We have been into tough times lately, but that does not erase what we have accomplished and the strong legacy we inherited to build on_".

 * __Present__: What are the most important challenges and opportunities?

 "_We are fighting to prove our worth to this organization. If we rise to that challenge, we will secure an influential place in the strategic planning process for the company going forward, far into the future_". 

 * __Future__: What are our inspiring ambitions?

 "_We can be the engine that drives our firm to industry leadership. Other firms have rebounded from downturns based on the next wave of innovation. That is what we will do, too. We are going to drive the future of this firm_".

# Purpose And Vision
---

You don't want your people to leave their best motives behind them when they come to work, or be motivated to go somewhere else. Give them a reason to be proud of what they do, and to know it's making a positive difference. Give them a shared purpose and a compelling vision.

Purpose and vision apply to all executive leaders because they answer the critically important why questions that go to the fundamental motivations people bring with them, or leave behind, when they come to work.

Effective executives know they must have excellent answers to these why questions:

* Why does this organization exist?
* Why is what we are doing worth my effort?
* Why will my work make a difference?

## Purpose

Purpose is a values-driven, integrity-based motive that people in the organization can rally around.

To make purpose as clear and concise as possible, talk with people about what they think are the best reasons the organization exists. What makes them proud to work here? What's the larger reason beyond their individual role that adds meaningful value to other people, such as your customers, or society as a whole? Add your own thinking and hone in on a sharp, clear articulation of the positive underlying human value that's being advanced because of what your organization is trying to do. That is your purpose.

## Vision

Vision is an inspiring view of the better future strived to create by acting on the shared purpose.

To create a compelling vision, you must:

* Collect meaningful improvements for the future
* Add own thoughts
* Distill to concise phrase

# Manage Your Image
---

As an executive leader, you need to inspire confidence, even with people who don't know you well. A crucial part of leadership is having people look at you and gain confidence, seeing how you react. If you're in control, they're in control, and if you're not, they're not, and they lose confidence in themselves and in you.

Here is how to do it:

1. Direct your thoughts: Your number one priority is always about advancing a shared purpose for greater good.  Focus your thoughts, not on your inner unease, but on what matters most for the group.

2. Direct your feelings toward determination to make progress on that number one priority: Practice channeling your difficult feelings toward focused determination. Make friends with that feeling. Focus determination is well suited for even the toughest situations. But focus determination to do what?

3. Direct your actions: Get determined to focus on that number one priority and to take the best actions to advance that priority. Actions include what you say and do including your body language and tone of voice. We can't see our own actions, so the first step to inspire more confidence with your behaviors is to understand the current impact of your actions.

# Energize And Empower
---

"_Nothing good happens with a product team unless they're excited_" - Scott Cook

Ensure everyone who interacts with you leaves more energized, not more depleted. Don't just encourage; empower. Tell them you're counting on them to figure this out or get this done or drive these results.

Don't micromanage. Micromanagers tend to remain managers, not leaders. Certainly not executive leaders. Leaders learn to let go. You don't scale work through others by controlling them, but by unleashing their potential.

Build them up, build their confidence, empower, give them power, power to make choices and stand accountable. Give them responsibility and challenge them to excel.

Don't win arguments with your people. Instead, energize and empower them.

# Encourage Personal Excellence
---

"_A leader is someone who sees more potential in others than they see in themselves_" - Stephen Covey

Executive leaders need to unleash the full potential of the people around them. For their current performance and their future growth. You want to be the strongest advocate for your people's best selves because if you don't believe in them, it's harder for them to believe in themselves and try to give their best effort.

Encourage individual excellence and you earn loyalty and commitment. Here is how to do it:

1. __Ask others what drives them__: What are the conditions under which you perform your absolute best? How can you create those conditions even more often? And what really drives you? How can you tap into that more often? Help them discover this about themselves and take charge of their situations even more and support those conditions and drives when you can.

2. __Identify and discuss their strenghts__: Let them know you see them and help them put those strengths into action even more and continue to grow them.

3. __Reinforce their growth attempts__: Give lots of positive reinforcement for their attempts to excel, to reach for more, and move out of their comfort zone. Whether they succeed or fall short, follow up with them to lock in their learning and turn insights into habits. Ask what are your key takeaways from that meeting? How will you apply those learnings over the next two weeks? Be specific.

4. __Encourage setting goals__: Encourage them to set their own ambitious goals for both their performance and their development. Also encourage them to take their own accountability plans with feedback partners and perhaps check-ins with you too. So it all starts with you seeing the potential others might miss in themselves, then keep going. Help them see it, act on it, and achieve it.

# Create Collaboration Opportunities
---

The best executive leaders obsess about how to get people to work together effectively at all levels of the organization.

Here is what to do to create more collaboration:

1. __Help out beyond your role__: Reach out to people beyond your direct tasks, ask about their priorities, offer to help and then follow through. You'll stand out in a positive way and set an example, showing others how to do it by doing it.

2. __Reach out for feedback__: Say, "_It's important to me that we work well together. What's going well, and what else do you need from me?_".

3. __Work on cross-functional initiatives__.

4. __Communicate the importance of collaboration__: Do it consistently, and highlight great examples of people doing it.

5. __Reward collaboration__.

6. __Support ad hoc opportunities for collaboration__: Encourage people to work together across boundaries in the organization.

# Develop Others
---

"_Every business is the people business. Nothing we do is more important than hiring and developing people. At the end of the day, you bet on people, not strategies_" - Lawrence Bossidy

To sustain success over time, executive leaders need to take a disciplined, strategic approach to attracting, developing and retaining talented people.

1. __Develop your people__: Give them feedback on areas of strength and improvement and encourage them to take charge of inviting regular feedback from their other stakeholders. Help them set priorities to grow their skills and abilities in ways that are motivating to them, aligned with their career goals, and are good for the organization. Shape their roles or add assignments that challenge them to expand their perspective, business acumen and skills. Give them something that requires more innovation or is larger in implementation scale or budget or covers wider geographic locations or has more deadline pressure or more people or more complexity and so forth. Help them find education and training opportunities aligned with their growth goals. Encourage and support them to build developmental relationships. Give them exposure to senior leaders, in action and as advice givers.

2. __Apply this across all the organization__: Research shows the most successful organizations over time implement a systematic approach to first, assess the performance and potential of their managers and leaders, second, identify areas of growth and development and third, provide opportunities and support growth and development in those areas. To attract and keep the best people, it's crucial this flows to every management level. Cascade this throughout the entire firm and evaluate and reward managers for how well they develop their people.

# Communicate To Motivate And Inspire
---

To thrive as a C-level executive, an individual needs to be a great communicator, a collaborator, and a strategic thinker.

When you speak as an executive, do no harm. Be thoughtful and intentional about what you say. Don't let words fly out of your mouth that will do damage to your team, your organization, or your reputation.

Ensure your words and motives have integrity. Speak thoughtfully and with positive purpose. You don't have to say everything on your mind. Choose what will best serve the people you want to lead. Choose what best helps their performance, growth, and morale.

Add inspiration to information. As an executive leader, you're not just dispensing data, passing instructions along, and doling out information. Given your role, you're constantly seen as the spokesperson for why questions. Why should we care? Why should we try harder? Why should we believe we can succeed? Speak to their motives. That's where their passion is, their concerns, their fears, their hopes. That's where people get aligned and engaged and do great things. Make your message matter. Communicate to motivate. Add inspiration to information.

# Communicate To Replicate
---

"_Managers create complexity. Leaders create simplicity_" - Jack Welch

A key limiting factor for rising leaders is speaking too much in the language of their functional specialty: engineering, finance, technical language, even business jargon. Insecure managers try to sound smart.

# Prioritization
---

"_If you have more than three priorities, then you don't have any_" - Jim Collins

"_Deciding what not to do is as important as deciding what to do. It's only be saying no that you can concentrate on the things that are truly important_" - Steve Jobs

"_The essence of strategy is choosing what not to do_" - Michael Porter

Executive leaders must establish priorities and focus effort. They need a mindset shift from doing things right to doing the right things. Be clear, be strong. Focus your people on what to do and what not to do. Be a persistent prioritizer and a top three leader. Consistently ask yourself and challenge others, "What are your top three priorities, in order?

Here's how to do it:

1. __Be a persistent prioritizer__: Activity alone isn't progress and moving forward on what matters less is time and effort lost. Real progress is priority-based progress. Separate the seemingly urgent from the truly important by prioritizing. Clarify what matters most in this meeting, project, or discussion. Then act accordingly.

2. __Be a top three leader__: The top three priorities should be clear and explicit at every level in the organization. Get aligned with your manager about what your three top priorities are. And for each of your direct reports, insist they set top three priorities for their team, for each team member, and for themselves. Ask your people what their top three priorities are and how they're doing on them? Ask what their team's top three priorities are and how the team is doing on each of them? If your people are unclear, coach them toward understanding the difference between urgent and important, between short-term distractions and strategic priorities.

# Accountability
---

Accountability is about progress on your shared purpose. It should be invigorating, satisfying, and a source of pride.

When done well, accountability isn't a chore, it's invigorating. The specifics of your situations will differ, but you can count on both planned and ad hoc opportunities to use invigorating accountability that emphasizes purpose, not punishment, and is motivating, not mechanical. As an executive leader, always make follow through a want to do.

Here's how to ensure accountability that gets the results you really want:

1. __Results over control__: Focus on achieving results, not controlling people. Follow-through discussions and check-ins shouldn't be about keeping people in line, but rather the great results people are reaching for.

2. __Use several types of meetings__: Tactical meetings, typically weekly or biweekly, to deal with current operational issues. Strategic meetings, typically every four to six weeks to address matters that concern progress on strategic priorities. Long-term planning meetings, typically every four to six months, often held away from the office to separate people mentally and physically from being distracted by day to day matters.

 At the end of each meeting, all three types of meetings, ensure you devote time to clarify action items. Who does what and by when. Set the expectation that in the next meeting, people will report on their action item progress. This way, disciplined accountability is seamlessly ongoing for everyone at every level, tactical, strategic, and long-term planning.

3. __Always connect accountability to shared purpose__: The more you connect people's efforts to shared purpose and compelling vision, the more energy they have.

# Influence Without Authority
---

"_The executive who keeps on doing what he has done successfully before (...) is almost bound to fail_" - Peter Drucker

To get things done like the best executives do, don't rely on the power and status of your position or mere rules and regulations. Even if you have authority, influence without authority. Use a forward-thinking influence plan that helps you get the best results without damaging relationships.

In every influence attempt, strive to make the relationship better because of how you conducted yourself and were thoughtful of the other's point of view.

Here is how to do it:

1. __Clarify your desired outcomes__
2. __Learn what influences the influencers__
3. __Choose the best methods__
4. __Invest in relationships__

# Cultivating Creativity
---

"_The most creative cientists are impelled by curiosity_" - Arthur Schalo

"_Creative inspiration does exist, but it must find you working_" - Picasso

Your work as an executive is to one: be impelled by curiosity, and two: encourage and reward innovation at all levels in your organization.

To enable organizations to survive and thrive, it's essential for executive leaders to cultivate creative thinking and innovation. You can even get creative about how you encourage creativity. Discuss creativity best practices and examples of innovation inside and outside your organization, and encourage others to do it too.

Here is how to do it:

1. __Add creative elements to roles__
2. __Reward success and failure__
3. __Make room for some rebellion__
4. __Provide training__
5. __Encourage other-focused innovation__
6. __Challenge teams to develop new products, services, and methods__
7. __Company-wide norms and practices__
8. __Dedicated creative time__
9. __Contests__

# Leading Change
---

"_If you want to make enemies, try to change something_" - Woodrow Wilson

Change is hard. Leading change is very hard, and executive leaders need to drive the most significant changes of all.

1. __Explain why change is necessary__
2. __Communicate a motivating change vision__
3. __Formulate a plan__
4. __Implement and learn__
5. __Celebrate the wins__
6. __Record what you learn__

# Self-Awareness
---

"_Be yourself, everyone else is taken_" - Oscar Wilde

Great leaders know their weaknesses and build relationships with people who have those qualities as strengths. They partner with them, have them as trusted advisors, and delegate to them.

Invest in your strengths, and relationship-build your weaknesses. You should also reflect on the types of situations that trigger you to reflexively respond with fight, flight, or freeze. You get disproportionately angry, that's fight, submissive, that's flight, or paralyzed, that's freeze, all of which can hurt your results and your reputation. Identify those situations which are unique to us, and then replace fight, flight, or freeze with purpose, determination, and adding value. Direct your thoughts towards the highest priority positive purpose relevant to that meeting or discussion.

Direct your feelings toward determination and your actions toward adding value.

1. __What are your top three defining moments__?

 Defining moments come in different forms. It might be words from a friend or a struggle you endured, failure that you recovered from or a triumphant achievement, a turning point when suddenly you see things differently.

2. __What are your top skills, abilities, or characteristics__?

 Defining moments often give you insight into your key leadership strengths. What are your top skills, abilities, or characteristics? Reflect on them and also get insights from different people who see you in action. We often miss some strengths because they feel natural and normal.

# Resilience
---

Psychology research shows that people who interpret setbacks as permanent, pervasive, and personal tend to have much more difficulty than those who do the opposite: seeing failures as temporary, specific, and external.

The pressures and demands of executive leadership can be draining and consuming, so find ways to fortify the four key elements of your well-being:

1. __Mental wellness__
2. __Emotional wellness__
3. __Physical wellness__
4. __Spiritual wellness__

# Learning
---

"_Every person I work with knows something better than me. My job is to listen long enough to find it and use it_" - Jack Nichols

"_Learn from other people's mistakes, you can't live long enough to make them all yourself_" - Eleanor Roosevelt

"_My biggest motivation is to keep challenging myself. Every day I'm learning something new_" - Richard Branson

Here is how to do it:

1. __Learn from experience__: Learning agility is the number one predictor of executive success. It can be summarized as the ability to know what to do when you can't know what to do. Successful executives need to learn at an extra level of challenge and velocity.

2. __Learn from anyone__: When you cross paths with people don't leave without learning something, maybe even about yourself.

3. __Learn at multiple levels__: A good way to ensure you're learning what you need for the next level of responsibility is to keep learning beyond your current role. You can learn about your business unit or function, your organization, your customers or clients, your industry, the competitive landscape, trends and technology, the economy, social changes, international affairs, government policies, and leadership skills, best practices, and other leader's successes. Do yourself a favor and learn from other leaders' failures too.

4. __Set learning priorities and implement learning plans__

# Connections
---

The best executive leaders keep connecting and building relationships with people, people senior to them, junior to them and with peers.

In your network over time fill all three levels with strong relationships of diverse people, people with different backgrounds, perspectives, ways of thinking, and people from different countries and cultures:

1. __People senior to you__: Connect with people who are older or more experienced than you and gain their insights and wisdom. I strongly recommend you build a personal board of advisors of four or more mentors who will advise, coach, encourage and stand up for you.

2. __People junior to you__: They are called "fat brains" because studies show they have more fat content in their brains than people at later ages. Everyone should have several fat brains in their lives.

3. __Peers__: Having a core group of supportive peers outside the politics of your own firm is a solid best practice. These are people who will push you in exchange for being pushed, who will raise the bar and tell you the truth.

Finally, for all three levels, be sure to connect with people who are unlike you. Keep connecting with seniors, juniors and peers, including some exact opposites who might be exactly what you need for executive success.
